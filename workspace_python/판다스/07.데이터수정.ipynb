```python
import pandas
df = pandas.read_excel('test1.xlsx')
df
```

```python
# 모든 학생의 영어점수를 현재 점수 + 5
df['영어점수'] = df['영어점수'] + 5
```

```python
# 현재 국어, 영어 점수를 통한 평균 변경
df['평균'] = (df['국어점수']+df['영어점수'])/2
df
```

```python
# 컬럼 추가하기
df['결과'] = 'Fail'
df
```

```python
# 컬럼 삭제하기
df.drop(columns={'결과'}) # 결과 컬럼 삭제
df.drop(columns={'평균','결과'}) # 평균,결과 컬럼 삭제
```

```python
# 특정 조건에 맞는 데이터만 수정
# 국어, 영어의 합이 170 점 이상이라는 조건
df.loc[df['국어점수']+df['영어점수'] >= 170,'결과']='Pass'
df
```

```python
# 학번에서 '번'이라는 글자를 제거
# 힌트 : 문자열은 리스트, int() 안에 문자열을 넣으면 숫자로 나옴


# DataFrame 에서 문자열 데이터를 처리할 때는 컬럼 데이터에 .str 이라는 키워드를 붙여서 사용 해야 함.
df['학번'] = df['학번'].str[:-1]

df['학번'] = df['학번'].str.replace('번','')
df
```
